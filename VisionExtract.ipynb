```python
import sys
from rembg import remove
from PIL import Image
import os

def process_image(input_path, output_path):
    """
    Processes an image to remove its background and saves the new image.
    The background is made black.
    """
    try:
        # Open the image file
        with open(input_path, 'rb') as i:
            with open(output_path, 'wb') as o:
                # Use rembg to remove the background.
                # The result is a new image with a transparent background.
                input_data = i.read()
                output_data = remove(input_data)

                # We save the output to the specified path
                o.write(output_data)

        # Convert the transparent background to black as per the project requirement
        convert_transparent_to_black(output_path)
        print(f"✅ Success: Background removed and converted to black. Output saved to '{output_path}'")

    except FileNotFoundError:
        print(f"❌ Error: The file '{input_path}' was not found. Please check the path and try again.")
    except Exception as e:
        print(f"❌ An unexpected error occurred: {e}")

def convert_transparent_to_black(image_path):
    """
    Converts a transparent background in a PNG image to a black background.
    """
    img = Image.open(image_path).convert('RGBA')
    # Create a new blank black image with the same size
    black_bg = Image.new('RGBA', img.size, (0, 0, 0, 255))
    # Composite the original image over the black background
    new_img = Image.alpha_composite(black_bg, img)
    # Convert back to RGB to remove the alpha channel before saving
    new_img_rgb = new_img.convert('RGB')
    new_img_rgb.save(image_path)

# Define input and output paths directly in the notebook
# Use the path from the downloaded dataset
dataset_path = "/kaggle/input/coco-2017-dataset"
input_image_path = os.path.join(dataset_path, "coco2017", "train2017", "000000000036.jpg") # Using a different example image
output_image_path = "output_image_black_bg_2.png" # New output file name

# Process the image
process_image(input_image_path, output_image_path)
```
